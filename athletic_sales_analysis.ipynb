```python

# Import Libraries and Dependencies
import pandas as pd
import numpy as np
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
#Import th CSVs
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020_df, sales_2021_df], axis='rows', join='inner').reset_index(drop=True)
combined_sales_df.head()
```

```python
# Check if any values are null.
if combined_sales_df.isnull().values.any():
    print('Data contains null values')
    
else: print('Data does not contain null values')

```

```python
# Check the data type of each column
combined_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df['invoice_date'] = pd.to_datetime(combined_sales_df['invoice_date'], format='%Y-%m-%d')
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the "units_sold" column to "Total_Products_Sold" 
sales_by_region_groupbymethod = combined_sales_df.groupby(['region', 'state', 'city']).agg(Total_Products_Sold = ('units_sold', 'sum')) 
sales_by_region_groupbymethod = sales_by_region_groupbymethod.sort_values(by='Total_Products_Sold', ascending=False)
sales_by_region_groupbymethod.head()
# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
sales_by_region = pd.pivot_table(combined_sales_df, index=['region', 'state', 'city'], values=['units_sold'], aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
sales_by_region.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
sales_by_region = sales_by_region.sort_values(by='units_sold', ascending=False)
sales_by_region.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total_Sales"
products_sold_by_region = combined_sales_df.groupby(['region', 'state', 'city']).agg(Total_Sales = ('total_sales', 'sum')) 
products_sold_by_region = products_sold_by_region.sort_values(by='Total_Sales', ascending=False)
products_sold_by_region.head()
# Show the top 5 results.
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
sales_by_date_table = pd.pivot_table(combined_sales_df, index=['invoice_date'], values=['total_sales'], aggfunc= 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
sales_by_date_table = sales_by_date_table.rename(columns={'total_sales':'Total_Sales'})
sales_by_date_table = sales_by_date_table.sort_values(by='Total_Sales', ascending=False) 
# Show the table.
sales_by_date_table.head(10)
#print ("The date with the highest highest sales is") + sales_by_date_table.max() + print ("and the date with the lowest sales is") + sales_by_date_table.min() 
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = sales_by_date_table.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales = daily_sales.sort_values (by='Total_Sales', ascending=False) 
daily_sales.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = sales_by_date_table.resample('W').sum()

# Sort the resampled pivot table in ascending order based on the existing column name.
weekly_sales = weekly_sales.sort_values(by='Total_Sales', ascending=False)
# Show the table.
weekly_sales.head(10)
```
